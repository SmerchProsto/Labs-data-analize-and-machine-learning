```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('company_sales_data.csv', delimiter=',')
```

```python
plt.figure()

plt.plot(df['month_number'], df['total_profit'])

plt.xticks(df['month_number'])
plt.title('Company profit per month')
plt.xlabel('Month number')
plt.ylabel('Total income')

plt.show()
```

```python
plt.figure()
plt.plot(df['month_number'], df['total_units'], 'r--', marker='o', label ='Profit data of last year')

plt.xticks(df['month_number'])
plt.title('Company Sales data of last year')
plt.legend(loc='lower right',  prop={'size': 8}, fontsize=12)
plt.xlabel('Month Number')
plt.ylabel('Total Sales')

plt.show()
```

```python
plt.figure()
# b - blue, g - green, r - red, c - , m - purple, y - yellow, k - black, w - white
plt.plot(df['month_number'], df['facecream'], 'b', marker='o', label ='Face cream sales data')
plt.plot(df['month_number'], df['facewash'], 'y', marker='o', label ='Face wash sales data')
plt.plot(df['month_number'], df['toothpaste'], 'g', marker='o', label ='Toothpaste sales data')
plt.plot(df['month_number'], df['bathingsoap'], 'r', marker='o', label ='Bathing soap sales data')
plt.plot(df['month_number'], df['shampoo'], 'm', marker='o', label ='Shampoo sales data')
plt.plot(df['month_number'], df['moisturizer'], 'k', marker='o', label ='Moisturizer sales data')

plt.xticks(df['month_number'])
plt.title('Sales data')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left', prop={'size': 8}, fontsize=12)

plt.show()
```

```python
plt.figure()
plt.plot(df['month_number'], df['toothpaste'], 'b', linestyle=' ', marker='o', label ='Toothpaste sales data')

plt.xticks(df['month_number'])
plt.title('Tooth paste Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Number of units sold')

plt.legend(loc='upper left',  prop={'size': 8}, fontsize=12)
plt.grid()

plt.show()
```

```python
fig, ax = plt.subplots()
ax.set_title('Face wash and Face cream sales data', fontsize=12)
ax.set_xlabel('Month, Number', fontsize=12)
ax.set_ylabel('Sales units in number', fontsize=12)
ax.set_xticks(df['month_number'])

ax.bar(df['month_number'] - 0.15, df['facecream'], width = 0.3, label='Face cream sales data')
ax.bar(df['month_number'] + 0.15, df['facewash'], width = 0.3, label='Face wash sales data')

ax.legend(loc='upper left', prop={'size': 8}, fontsize = 12)
ax.grid(linestyle='--')

plt.show()
```

```python
fig, ax = plt.subplots()
x = df['month_number']
y = df['bathingsoap']
ax.bar(x, y)

ax.set_title('Bathing soap sales data')
ax.set_xlabel('Month Number')
ax.set_ylabel('Sales units in number')
ax.set_xticks(df['month_number'])
ax.grid(linestyle='--')

plt.show()
```

```python
fig, ax = plt.subplots()
ax.set_title('Profit data', fontsize=12)
ax.set_xlabel('Profit range in dollar', fontsize=12)
ax.set_ylabel('Actual profit in dollar', fontsize=12)

ax.hist(df['total_profit'], bins=4, label='Profit data', range=[150000, 250000], color='blue')
ax.hist(df['total_profit'], bins=2, range=[250000, 350000], color='blue')

ax.legend(loc='upper right', prop={'size': 8}, fontsize = 12)

plt.show()
```

```python
labels = ['Face cream', 'Face wash', 'Tooth paste', 'Bathing soap', 'Shampoo', 'Moisturizer']
values = [sum(df['facecream']) / sum(df['total_units']), sum(df['facewash']) / sum(df['total_units']), sum(df['toothpaste']) / sum(df['total_units']), sum(df['bathingsoap']) / sum(df['total_units']), sum(df['shampoo']) / sum(df['total_units']), sum(df['moisturizer']) / sum(df['total_units'])]
fig1, ax1 = plt.subplots()
ax1.pie(values, labels=labels, autopct='%1.1f%%')
plt.legend(fontsize=10, loc=4)
plt.show()
```

```python
fig, ax = plt.subplots(2)
ax[0].plot(df['bathingsoap'], 'k', marker='o')
ax[1].plot(df['month_number'], df['facewash'], 'r', marker='o')

ax[0].set_title('Sales data of Bathing soap')
ax[1].set_title('Sales data of Face wash')
ax[1].set_xlabel('Month Number')
ax[1].set_ylabel('Sales units in number')

ax[0].set_xticks([])
ax[1].set_xticks(df['month_number'])

plt.show()
```

```python
fig, ax = plt.subplots()
x = df['month_number']
y1 = df['facecream']
y2 = df['facewash']
y3 = df['toothpaste']
y4 = df['bathingsoap']
y5 = df['shampoo']
y6 = df['moisturizer']

ax.stackplot(x,[y1, y2, y3, y4, y5, y6])

ax.set_title('All product sales data using stack  plot')
ax.set_xlabel('Month Number')
ax.set_ylabel('Sales units in Number')
ax.legend(loc='upper left', prop={'size': 8}, fontsize=12)
plt.show()
```
